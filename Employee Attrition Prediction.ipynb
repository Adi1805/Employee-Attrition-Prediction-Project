```python
X_cat = df[['BusinessTravel', 'Department', 'EducationField', 
                     'Gender', 'JobRole', 'MaritalStatus']]
X_cat
```

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
X_cat = onehotencoder.fit_transform(X_cat).toarray()
```

```python
X_cat.shape
```

```python
X_cat = pd.DataFrame(X_cat)  # Corrected 'Dataframe' to 'DataFrame' - pandas function names are case-sensitive
```

```python
X_cat
```

```python
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
#Selecting all numerical columns from the dataset except the column Attrition
X_numerical = df[numerical_columns].drop('Attrition', axis=1)
X_numerical
```

```python
X_all = pd.concat([X_cat, X_numerical], axis = 1)
X_all
```

```python
from sklearn.preprocessing import MinMaxScaler

# Converting all column names to strings before scaling
X_all.columns = X_all.columns.astype(str)

# Now apply the scaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X_all.values)
```

```python
X
```

```python
y=df["Attrition"]#target Variable
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)#Splitting the model 
#into testing and training
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.impute import SimpleImputer
import numpy as np

# Creating an imputer to replace NaN values with the mean of each column
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

# Fit the imputer on X_train and transform both X_train and X_test
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

#Training the Logistic Regression Model
model = LogisticRegression()
model.fit(X_train_imputed, y_train)

#Making Predictions
y_pred = model.predict(X_test_imputed)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Correct order of arguments: true labels first, then predictions
print("Logistic Regression Accuracy: {} %".format(100 * accuracy_score(y_test, y_pred)))
```

```python
from sklearn.metrics import classification_report

print("--- Logistic Regression Classification Report ---")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier

# Create and train the Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train_imputed, y_train)

# Make predictions
y_pred_dt = dt_model.predict(X_test_imputed)

# Print accuracy
print("Decision Tree Accuracy: {} %".format(100 * accuracy_score(y_test, y_pred_dt)))
```

```python
print("\n--- Decision Tree Classification Report ---")
print(classification_report(y_test, y_pred_dt))
```

```python
from sklearn.svm import SVC

# Create and train the SVM model
svm_model = SVC()
svm_model.fit(X_train_imputed, y_train)

# Make predictions
y_pred_svm = svm_model.predict(X_test_imputed)

# Print accuracy
print("SVM Accuracy: {} %".format(100 * accuracy_score(y_test, y_pred_svm)))
```

```python
print("\n--- Support Vector Machine (SVM) Classification Report ---")
print(classification_report(y_test, y_pred_svm))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create and train the Random Forest model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train_imputed, y_train)

# Make predictions
y_pred_rf = rf_model.predict(X_test_imputed)

# Print accuracy
print("Random Forest Accuracy: {} %".format(100 * accuracy_score(y_test, y_pred_rf)))
```

```python
print("\n--- Random Forest Classification Report ---")
print(classification_report(y_test, y_pred_rf))
```
